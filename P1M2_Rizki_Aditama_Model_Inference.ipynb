# X. MODEL INFERENCE

```python
import numpy as np
import pickle
import random
import pandas as pd
```

```python
# Load All Files
with open('all_process_rizki.pkl', 'rb') as file_1:
  all_process = pickle.load(file_1)
```

```python
# Membuat list kosong untuk menyimpan data inference
data_inference_list = []

# Jumlah data inference yang ingin Anda buat (dalam kasus ini 100 baris)
num_samples = 1000

# Batasan nilai minimum dan maksimum untuk kolom "tenure"
min_tenure = 1
max_tenure = 72
# Kolom "tenure"
tenure = np.random.uniform(min_tenure, max_tenure, num_samples)

# Batasan nilai minimum dan maksimum untuk kolom "MonthlyCharges"
min_mc = 19
max_mc = 72
# Kolom "MonthlyCharges"
MonthlyCharges = np.random.uniform(min_mc, max_mc, num_samples)  

# Kolom "Contract"
Contract = np.array(['Month-to-month', 'One year', 'Two year'])

# Kolom "OnlineSecurity"
OnlineSecurity = np.array(['No', 'No internet service', 'Yes'])

# Kolom "TechSupport"
TechSupport = np.array(['No', 'No internet service', 'Yes'])

# Kolom "PaymentMethod"
PaymentMethod = np.array(['Electronic check', 'Mailed check', 'Bank transfer (automatic)','Credit card (automatic)'])

# Kolom "InternetService"
InternetService = np.array(['No', 'DSL', 'Fiber optic'])

```

```python
# Loop untuk membuat data inference sebanyak num_samples
for _ in range(num_samples):
    random_tenure_value = np.random.choice(tenure)
    random_MonthlyCharges_value = np.random.choice(MonthlyCharges)
    random_Contract_value = np.random.choice(Contract)
    random_OnlineSecurity_value = np.random.choice(OnlineSecurity)
    random_TechSupport_value = np.random.choice(TechSupport)
    random_PaymentMethod_value = np.random.choice(PaymentMethod)
    random_InternetService_value = np.random.choice(InternetService)
    
    data_inference_list.append([random_tenure_value, MonthlyCharges[_], random_Contract_value,
                                random_OnlineSecurity_value, random_TechSupport_value,
                                random_PaymentMethod_value,random_InternetService_value])

# Membuat DataFrame data inference
data_inference = pd.DataFrame(data_inference_list, columns=['tenure', 'MonthlyCharges', 'Contract',
                                                           'OnlineSecurity', 'TechSupport', 'PaymentMethod', 'InternetService'])

# Menampilkan data inference
data_inference
```

```python
data_inference['InternetService'] = data_inference['InternetService'].replace({'No' : 0, 'DSL': 1, 'Fiber optic':2})
data_inference['OnlineSecurity'] = data_inference['OnlineSecurity'].replace({'No' : 0, 'No internet service': 1,'Yes' : 2 })
data_inference['TechSupport'] = data_inference['TechSupport'].replace({'No' : 0, 'No internet service': 1,'Yes' : 2 })
data_inference['Contract'] = data_inference['Contract'].replace({'Month-to-month' : 0, 'One year': 1,'Two year' : 2 })
data_inference['PaymentMethod'] = data_inference['PaymentMethod'].replace({'Electronic check' : 0, 'Mailed check' : 1, 'Bank transfer (automatic)': 2, 'Credit card (automatic)' : 3  })
data_inference.head(5)
```

```python
y_pred_inf = all_process.predict(data_inference)
y_pred_inf
```

```python
# Convert into DataFrame
y_pred_inf = pd.DataFrame(y_pred_inf, columns=['Churn - Prediction'])
y_pred_inf
```

```python
# Merge Inference-Set and Prediction
tabel_inference = pd.concat([data_inference, y_pred_inf], axis=1)
tabel_inference
```

```python
# Mengganti nilai 'Dropout' dengan 0 dan 'Graduate' dengan 1 pada kolom 'Target'
tabel_inference['Churn - Prediction'] = tabel_inference['Churn - Prediction'].replace({0 : 'Churn', 1 : 'Loyal'})
tabel_inference
```
